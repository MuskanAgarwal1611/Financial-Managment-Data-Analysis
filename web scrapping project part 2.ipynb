# web scrapping

### importing libraries

```python
import numpy as np
import pandas as pd
```

### reading file

```python
m=pd.read_html(r'https://en.wikipedia.org/wiki/List_of_territorial_entities_where_English_is_an_official_language')
m
```

```python
m1=m[0]
m1
```

```python
m2=m[1]
m2
```

```python
country1=m1.loc[:,['Alpha-3 code','Primary language?']]
country1
```

```python
country2=m2.loc[:,['Alpha-3 code','Primary language?']]
country2
```

### concatenating two tables

```python
master=pd.concat([country1,country2],axis=0)
```

```python
master.head()
```

```python
master.reset_index(drop='index',inplace=False)
master.head()
```

### renaming columns

```python
master.rename( columns={'Alpha-3 code':'country_code','Primary language?':'official language'},inplace=True)
```

```python
master.shape
```

```python
master['official language']=master['official language'].apply(lambda x:x.split()[0])
```

```python
master.head()
```

```python
master.to_csv('english_countries.csv',index=False)
```
